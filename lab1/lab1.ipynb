```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

images = np.zeros((9, 400, 600))
for i in range(9):
    image = np.load(f'images/car_{i}.npy')
    images[i] = image
    io.imshow(image.astype(np.uint8))
    io.show()
```

```python
mean_image = np.mean(images, 0)

io.imshow(mean_image.astype(np.uint8))
io.show()
```

```python
# f
deviatia_standard = np.std(images)
print(deviatia_standard)
```

```python
# g
images2 = np.zeros((9, 400, 600))
for i in range(9):
    images2[i] = (images[i] - mean_image) / deviatia_standard
    io.imshow(images2[i].astype(np.uint8))
    io.show()
```

```python
# h

images3 = np.zeros((9, 101, 121))
for i in range(9):
    images3[i] = images[i][200:301, 280:401]
    io.imshow(images3[i].astype(np.uint8))
    io.show()
    
```
